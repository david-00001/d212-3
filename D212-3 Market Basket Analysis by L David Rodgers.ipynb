```python
# Import packages
import pandas as pd
```

```python
import numpy
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Read in CSV to dataframe
df = pd.read_csv('teleco_market_basket.csv')
```

```python
# Initial exploration of dataset
df.info()
```

```python
# Drop rows containing only null values
df = df.dropna(how='all')

# Fill in remaining null values
df = df.fillna('')
```

```python
# Convert df to list format (GeeksforGeeks, 2023)
transactions = df.values.tolist()
print(transactions[0])
```

```python
# Create a TransactionEncoder object and fit the transactions
encoder = TransactionEncoder().fit(transactions)

# One-hot encode transactions
onehot = encoder.transform(transactions)

# Convert encoded data to dataframe
onehot = pd.DataFrame(onehot, columns = encoder.columns_)

# Drop the column that represents empty items
onehot = onehot.drop(columns='')

# Print first five rows of new dataframe
onehot.head()
```

```python
# Apply the Apriori algorithm
frequent_itemsets = apriori(onehot, min_support = 0.01, use_colnames=True)
print(f'How many itemsets that met minimum support: {len(frequent_itemsets)}')
frequent_itemsets.head()
```

```python
# Generate the association rules
rules = association_rules(frequent_itemsets, metric='support', min_threshold=0.0)
print(len(rules))
rules.head()
```

```python
import seaborn as sns

sns.scatterplot(x='antecedent support', y='consequent support', data=rules)
```
